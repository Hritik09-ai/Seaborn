```python
# QUESTION.1 Name any five plots that we can plot using the Seaborn library. Also, state the uses of each plot.
# ANSWER 
# Seaborn is a powerful Python data visualization library built on top of Matplotlib. It provides a high-level
# interface for creating informative and attractive statistical graphics. Here are five common plots that you can
# create using Seaborn along with their typical use cases:

# Scatter Plot:
# Use: Scatter plots are used to visualize the relationship between two continuous variables. They are great for
# identifying patterns, trends, or clusters in data points.
# Seaborn Function: seaborn.scatterplot()

# Bar Plot:
# Use: Bar plots are used to display and compare the values of categorical data. They are suitable for showing
# the distribution of a single categorical variable or comparing multiple categories.
# Seaborn Function: seaborn.barplot()

# Histogram:
# Use: Histograms are used to visualize the distribution of a single continuous variable. They help you understand
# the frequency or probability distribution of data.
# Seaborn Function: seaborn.histplot()

# Box Plot (Box and Whisker Plot):
# Use: Box plots are useful for displaying the distribution of data and identifying outliers. They show the median,
# quartiles, and potential outliers in a dataset.
# Seaborn Function: seaborn.boxplot()

# Heatmap:
# Use: Heatmaps are great for visualizing relationships and patterns in a matrix or 2D dataset. They are often used
# in correlation matrices and to display data with color intensity.
```

```python
# QUESTION.2  Load the "fmri" dataset using the load_dataset function of seaborn. Plot a line plot using
#                x = "timepoint" and y = "signal" for different events and regions.
import seaborn as sns
import matplotlib.pyplot as plt

# Load the "fmri" dataset using seaborn
fmri_data = sns.load_dataset("fmri")

# Create a line plot using seaborn
plt.figure(figsize=(6, 4))  # Adjust the figure size as needed

# Specify the x-axis and y-axis variables, and use 'hue' for different events and regions
sns.lineplot(data=fmri_data, x="timepoint", y="signal", hue="event", style="region")

# Add labels and a legend
plt.xlabel("Timepoint")
plt.ylabel("Signal")
plt.title("FMRI Signal over Time by Event and Region")
plt.legend(title="Event")

# Show the plot
plt.show()
```

```python
# QUESTION.3 Load the "titanic" dataset using the load_dataset function of seaborn. Plot two box plots
#            using x = 'pclass', y = 'age' and y = 'fare'.
# ANSWER 
import seaborn as sns
# Load the Titanic dataset
titanic_data = sns.load_dataset("titanic")

import matplotlib.pyplot as plt
# Create the first box plot: 'pclass' vs. 'age'
plt.figure(figsize=(6, 4))
sns.boxplot(x='pclass', y='age', data=titanic_data)
plt.title('Box Plot of Age by Passenger Class')
plt.xlabel('Passenger Class')
plt.ylabel('Age')
plt.show()

# Create the second box plot: 'pclass' vs. 'fare'
plt.figure(figsize=(6, 4))
sns.boxplot(x='pclass', y='fare', data=titanic_data)
plt.title('Box Plot of Fare by Passenger Class')
plt.xlabel('Passenger Class')
plt.ylabel('Fare')
plt.show()
```

```python
# QUESTION.4 Use the "diamonds" dataset from seaborn to plot a histogram for the 'price' column. Use the hue
#        parameter for the 'cut' column of the diamonds dataset.
# ANSWER 
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset
diamonds = sns.load_dataset('diamonds')

# Create a histogram with hue for 'cut' column
sns.histplot(data=diamonds, x='price', hue='cut', kde=True)

# Add labels and title
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.title('Histogram of Diamond Prices by Cut')

# Show the plot
plt.show()
```

```python
# QUESTION.5 Use the "iris" dataset from seaborn to plot a pair plot. Use the hue parameter for the "species" column
#          of the iris dataset.
# ANSWER 
import seaborn as sns
import matplotlib.pyplot as plt

# Load the iris dataset
iris = sns.load_dataset("iris")

# Create a pair plot with the "species" column as the hue parameter
sns.pairplot(iris, hue="species")

# Show the plot
plt.show()
```

```python
# QUESTION.6 Use the "flights" dataset from seaborn to plot a heatmap.
# ANSWER 
import seaborn as sns
import matplotlib.pyplot as plt

# Load the flights dataset (assuming you have Seaborn and Matplotlib installed)
flights_data = sns.load_dataset("flights")

# Pivot the data to create a pivot table for the heatmap
pivot_data = flights_data.pivot("month", "year", "passengers")

# Create a heatmap using Seaborn
plt.figure(figsize=(10, 6))
sns.heatmap(pivot_data, annot=True, fmt="d", cmap="YlGnBu", linewidths=0.1)
plt.title("Passenger Count by Year and Month")
plt.show()
```
